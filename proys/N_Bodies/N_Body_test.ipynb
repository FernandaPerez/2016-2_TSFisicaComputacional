```julia
# lista de planetas, posici√≥n x,y,x, velocidad x,y,z, masa
planet = (
[0,0,100,-1,0,0,.001],
[10,0,100,-1,.005,0,.0005],
#[150,150,0,0,1,0,50],

[0,0,0,.01,0,0,1]
)

# esto es para guardar las posiciones en cada tiempo
lista=Array(Array,0)
for i=1:length(planet)
    array=Array(Array,3)
    push!(lista,array)
    array[1]=Array(Number,0)
    array[2]=Array(Number,0)
    array[3]=Array(Number,0)
end
    
for fr=1:2000
    for i=1:length(planet)
        for ii=1:length(planet)
            if(i!=ii)
                rx=planet[ii][1]-planet[i][1]
                ry=planet[ii][2]-planet[i][2]
                rz=planet[ii][3]-planet[i][3]

                r2=(rx*rx)+(ry*ry)+(rz*rz)
                r=sqrt(r2)
                a=(planet[ii][7])/(r2)
                p=a/r

                ax=a*rx
                ay=a*ry
                az=a*rz

                planet[i][4]+=ax
                planet[i][5]+=ay
                planet[i][6]+=az
            end
        end
        
        planet[i][1]+=planet[i][4]
        planet[i][2]+=planet[i][5]
        planet[i][3]+=planet[i][6]
    end
    for i=1:length(planet)
        push!(lista[i][1],planet[i][1])
        push!(lista[i][2],planet[i][2])
        push!(lista[i][3],planet[i][3])
    end
end
```
